**Adding the total number of reactions**

```python
actual_reactions_list = ["fb_love", "fb_haha", "fb_wow", "fb_angry", "fb_sad"]
reactions_list = ["fb_like"] + actual_reactions_list

def fb_total_reactions(dataframe_row):
    return sum(dataframe_row[r] for r in reactions_list)

function = fb_total_reactions
dataset.loc[:,function.__name__] = dataset.apply(lambda x: function(x), axis=1)
```

## Visualization

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
%matplotlib inline
sns.set(style='whitegrid', rc={"grid.linewidth": 0.1})
sns.set_context("paper")

plt.rc('figure', figsize=(8,4))
plt.rc('figure.subplot', wspace=.33)
```

```python
import numpy as np

def reduce_ticks(num, n_ticks = 3):
    num /= 2
    div = 1
    while num > 9:
        num /= 10
        div *= 10
    return num * div

def popularity_fb_metrics(df, title="", figsize=(15, 8), all_data=True):
    fig = plt.figure(figsize=figsize)
    plt.title(title, fontsize=18)
    ts = pd.DataFrame(data = df[["fb_total_reactions","num_comments","shares"]])
    ts.index = pd.to_datetime(df["time_created"])
    if all_data:
        reactions_ax = fig.add_subplot(311)
    else:
        reactions_ax = fig.add_subplot(111)
    ts.fb_total_reactions.plot(x="time_created", y="fb_total_reactions", ax=reactions_ax)
    reactions_ax.set_ylabel('Reactions', fontsize=16)
    reactions_ax.set_xlabel('Date', fontsize=16)
    maxval = max(ts.fb_total_reactions); ticks = np.arange(0, maxval, reduce_ticks(maxval))
    reactions_ax.set_yticks(ticks)
    
    if all_data:
        comments_ax = fig.add_subplot(312)
        shares_ax = fig.add_subplot(313)
        ts.num_comments.plot(ax=comments_ax)
        ts.shares.plot(ax=shares_ax)
        comments_ax.set_ylabel('Comments', fontsize=16)
        comments_ax.set_xlabel('Date', fontsize=16)
        shares_ax.set_ylabel('Shares', fontsize=16)
        shares_ax.set_xlabel('Date', fontsize=16)
        maxval = max(ts.num_comments); ticks = np.arange(0, maxval, reduce_ticks(maxval))
        comments_ax.set_yticks(ticks)
        maxval = max(ts.shares); ticks = np.arange(0, maxval, reduce_ticks(maxval))
        shares_ax.set_yticks(ticks)
    fig.set_tight_layout(True)
    plt.show()
    
for g, data in dataset.groupby(["database"]):
    popularity_fb_metrics(data, getlabel(g), figsize=(15, 3), all_data=False)
```

**Setting the most voted reaction (excluding "like") as a label.**

```python
reaction_class = {
    "fb_love": 0, 
    "fb_haha": 1,
    "fb_wow": 2,
    "fb_angry": 3,
    "fb_sad": 4
}

def highest_reaction(dataframe_row):
    reactions_row = dataframe_row[actual_reactions_list]
    reactions_row.sort_values(ascending=False, inplace=True)
    total_reactions = reactions_row.sum()
    
    max_reaction = reactions_row.index[0]
    max_votes = reactions_row.max()
    reaction_id = reaction_class[max_reaction]
    if max_votes != 0 and max_votes >= (total_reactions * 0.75):
        return reaction_id
    
    # if max_reaction == None:
    #    print("NA found")
    return None

function = highest_reaction
dataset_sliced.loc[:,function.__name__] = dataset_sliced.apply(lambda x: function(x), axis=1)
```

```python
dataset_sliced.head(2)
```

**And remove rows with no reaction or with number of votes less than 500**

```python
def remove_no_reaction(df):
    tmp = df[(df['highest_reaction'].notnull())]
    return tmp[tmp['fb_total_reactions'] >= 200]
    
dataset_sliced = remove_no_reaction(dataset_sliced)
dataset_sliced.loc[:,"highest_reaction"] = dataset_sliced.loc[:,"highest_reaction"].astype(int)
```

**If the 2 most voted reactions have more than 0.5% of the total reaction votes then set a label to it** (also excluding "like")

```python
extended_reaction = {}
import itertools
comb_count = 5
for comb in itertools.combinations(reaction_class, 2):
    comb = sorted(comb, reverse=False)
    extended_reaction[comb[0] + getlabel(comb[1])] = comb_count
    comb_count += 1
extended_reaction
```

```python
extended_reactions_list = actual_reactions_list + list(extended_reaction.keys())
extended_reactions_list
```

```python
def highest_reaction_extended(dataframe_row):
    max_votes = -1
    max_reaction = None
    reactions_row = dataframe_row[actual_reactions_list]
    reactions_row.sort_values(ascending=False, inplace=True)
    total_reactions = reactions_row.sum()
    tmp = sorted((reactions_row.index[0], reactions_row.index[1]), reverse=False)
    
    output = extended_reaction[tmp[0] + getlabel(tmp[1])]    
    # if reactions_row[0] >= (total_reactions * 0.75):
    #     output = reaction_class[reactions_row.index[0]]
    return output

function = highest_reaction_extended
dataset_sliced.loc[:,function.__name__] = dataset_sliced.apply(lambda x: function(x), axis=1)
```

## Finding correlations between likes and each reaction

The first interesting question I wanted to answer was is there a linear correlation between the likes and the reaction?, if there is not, then there may be some of them more correlated than the others and those should be the most important.

First by using linear regressions.

```python
def regplot(X, Y, xlabel="", ylabel="", title="", angle=0, fit_reg=False):
    plt.scatter(X, Y, color='#0080ff', edgecolor='black', linewidth='0.5')
    sns.regplot(X, Y, scatter=False,  fit_reg=fit_reg, color="#df2020")
    plt.ylim([0, 16000])
    plt.xlim([0, X.max()])
    if xlabel != "" and ylabel != "" and title == "":
        title = "%s vs %s" % (xlabel, ylabel)
    plt.title(title, fontsize=15, y=1.05)
    plt.xlabel(xlabel, fontsize=15)
    plt.ylabel(ylabel, fontsize=15)
    plt.xticks(rotation=angle)
    plt.tight_layout()
    
fig = plt.figure(figsize=(15,4))
plt.subplot(131)
regplot(dataset_sliced["fb_total_reactions"], dataset_sliced["fb_love"],
        getlabel("fb_love"), getlabel("fb_total_reactions").title(), angle=70, fit_reg=True)
plt.subplot(132)
regplot(dataset_sliced["fb_total_reactions"], dataset_sliced["fb_wow"],
        getlabel("fb_wow"), getlabel("fb_total_reactions").title(), angle=70, fit_reg=True)
plt.subplot(133)
regplot(dataset_sliced["fb_total_reactions"], dataset_sliced["fb_sad"],
        getlabel("fb_sad"), getlabel("fb_total_reactions").title(), angle=70, fit_reg=True)
```

## Text cleaning

We also figured out the dataset from "buzzfeed" contains english and spanish headlines, spanish ones must be taken away from the corpus. **I'm gonna include buzzfeed's data on this, it's the one that posts articles on different languages than English so I'll remove those, even if they are actually in English**.

```python
from langdetect import DetectorFactory
from langdetect import detect_langs
DetectorFactory.seed = 0  # It's non-deterministic

def getlanguage(text):
    if text == "":
        return None
    langs = detect_langs(text)
    if "en" in [i.lang for i in langs]:
        return "en"
    return langs[0].lang

print(getlanguage("17 Veces que la PolicÃ­a de Ecuador fue la reina de Twitter"))
print(getlanguage("An Adult's Guide To PokÃ©mon Go An Adult's Guide To PokÃ©mon Go"))
print(getlanguage("An Adult's Guide To PokÃ©mon Go"))
print(getlanguage(""))
```

**We are gonna have  headlines as features and the first thing to do is pre-processing the text, text cleaning.**

It is basically a sentiment analysis problem because sentiments/reactions are considered synonims and in that case, expressions like "22", "2.2", "f\*ck", "r.i.p.", "semi-final", "don't", "how?", "!!!" or ":-(", must be taken into account. Every word/expression should count, we can't even remove swear words.

```python
import nltk
from nltk.tokenize import RegexpTokenizer
pattern = r"""(?x)                            # set flag to allow verbose regexps
              (?:[a-zA-Z]\.)+                 # abbreviations, e.g. U.S.A.
              |[a-zA-Z]+(?:[-'*][a-zA-Z]+)*   # words w/ optional internal hyphens/apostrophe
            """
tokenizer = nltk.RegexpTokenizer(pattern)
```

```python
tokenizer.tokenize("22, 2.2, f*ck, r.i.p., semi-final, | don't, how?, !!! or :-(")
```

```python
tokenizer.tokenize('This Shelter Takes Dogs Out For â€˜Puppuccinosâ€™ To Find Them New Homes')
```

For now, we are gonna just avoid having punctuations and numbers and also first unidecode the input text.

```python
from unidecode import unidecode
unidecode("PokÃ¨mon, MontrÃ©al, Ã¼ber, 12.89, MÃ¨re, FranÃ§oise, noÃ«l, 889")
```

Some headlines have the form

"`<title> | <company_name>`" or "`<title> (by <authot>)`"

Those words do not describe what the post is about, so we remove the text after the "|" and everything between "(parentheses)".

```python
import re

def special(text):
    text = text.split("|")[0]
    text = re.sub("[\(].*?[\)]", "", text)
    return text

print(special("<title> | <company_name>"))
print(special("<title> (by <authot>)"))
```

Then, we add a preprocessed text column.

```python
def tokenize(text):
    text = text.lower()
    # Tokenize words without punctuation
    tokenizer = nltk.RegexpTokenizer(pattern)
    return tokenizer.tokenize(text)

def preprocessed_name(dataframe_row):
    return " ".join(tokenize(unidecode(special(dataframe_row["name"]))))

function = preprocessed_name
dataset_model.loc[:,function.__name__] = dataset_model.apply(lambda x: function(x), axis=1)
```

**Profanity censorship**

I'm not going to filter bad words, but only censor them cuz they are censored in most of the headlines. It is done for having more common words.

```python
bad_words_path = "resources/bad_words.json"  
with open(bad_words_path, "r") as file:
    bad_words = json.loads(file.read())
        
def censor(word):
    for bad_word in bad_words.keys():
        if bad_word in word:
            #word = word.replace(bad_word, bad_words[bad_word], 1)
            word = bad_words[bad_word]
    return word

def tokenize_bad_words(text):
    tokens = tokenize(text)
    tokens = [censor(word) for word in tokens]    
    return tokens

def preprocessed_name(dataframe_row):
    return " ".join(tokenize_bad_words(dataframe_row["preprocessed_name"]))

function = preprocessed_name
dataset_model.loc[:,function.__name__] = dataset_model.apply(lambda x: function(x), axis=1)
```

**Separating contractions**

```python
# Modified list of contractions gotten from this StackOverflow question:
# http://stackoverflow.com/questions/19790188/expanding-english-language-contractions-in-python
contractions_path = "resources/contractions.json"  
with open(contractions_path, "r") as file:
    contractions = json.loads(file.read())
    
def separate_contraction(word):
    for contraction in contractions.keys():
        if contraction in word:
            #word = word.replace(bad_word, bad_words[bad_word], 1)
            word = contractions[contraction]
    return word.split()

def tokenize_contractions(text):
    tokens = tokenize(text)
    output = []
    for word in tokens:
        output.extend(separate_contraction(word))
    return output

tokenize_contractions("that'll don't he'll gonna you're")
```

**Lemmatizing, stemming, stopword removal, separate contractions**

Because most of the time these steps are necessary.

```python
stopwords = nltk.corpus.stopwords.words('english')
special_stopwords_path = "resources/special_stopwords.txt"  
with open(special_stopwords_path, "r") as file:
    special_stopwords = file.read().splitlines()

from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

def tokenize_stem_stopwords(text):
    tokens = tokenize_contractions(text)
    tokens = [word for word in tokens if (word not in stopwords and word not in special_stopwords)]
    short_tokens = [wordnet_lemmatizer.lemmatize(t) for t in tokens]
    short_tokens = [stemmer.stem(t) for t in short_tokens]
    return short_tokens
```

```python
tokenize_stem_stopwords("that'll v don't you're take co-workers / co-worker taken r.i.p. probable. throne's. throne")
```

```python
def preprocessed_stem_stop(dataframe_row):
    return " ".join(tokenize_stem_stopwords(dataframe_row["preprocessed_name"]))

function = preprocessed_stem_stop
dataset_model.loc[:,function.__name__] = dataset_model.apply(lambda x: function(x), axis=1)
```
